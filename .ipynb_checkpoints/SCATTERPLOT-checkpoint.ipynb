```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_census = pd.read_csv(r'C:\Users\LENOVO\Desktop\DATA VISUALIZATION PROJECT\census.csv')
```

```python
df_census
```

```python
high_literacy=df_census.sort_values(by = 
"Literacy Rate",ascending = False).head(10)
name = np.arange(len(high_literacy["State/UT"]))
plt.figure(figsize=(15,8))
plt.scatter(x = high_literacy["State/UT"],y=high_literacy[
"Literacy Rate"],c='black')
plt.plot(high_literacy["State/UT"],high_literacy["Literacy Rate"],c='r')
plt.xlabel("Name of State")
plt.ylabel("Literacy Rate")
plt.title("Top 10 states with highest No Literacy Rate")
plt.grid()
plt.show()
```
